```julia
using Plots

N = 10
L = 0.1
T_0 = 10.0
t_1 = 0.25; t_2 = 1.0; t_3 = 10.0
a = 1.5 * 10^(-6)
T_1 = zeros(1000, 1)
T_2 = zeros(1000, 1)
T_3 = zeros(1000, 1)

function Temp(L::Float64, T_0::Float64, t::Float64, a::Float64, x::Float64)
    T_xt = 5 * exp(-a * t * 3600 * (3 * pi / L)^2) * sin(3 * pi * x / L) + T_0
end

for i = 1:1000
    T_1[i] = Temp(L, T_0, t_1, a, (i-1)/10000)
    T_1
end

for i = 1:1000
    T_2[i] = Temp(L, T_0, t_2, a, (i-1)/10000)
    T_2
end

for i = 1:1000
    T_3[i] = Temp(L, T_0, t_3, a, (i-1)/10000)
    T_3
end

x = 1:1000
```

График при t = 0,25 ч

```julia
plot((x .- 1) / 100, T_1, label="t = 0.25 h")
xlabel!("x, cm")
ylabel!("T(x), °C")
```

График при t = 1 ч

```julia
plot((x .- 1) / 100, T_2,  label="t = 1 h")
xlabel!("x, cm")
ylabel!("T(x), °C")
```

График при t = 10 ч

```julia
plot((x .- 1) / 100, T_3,  label="t = 10 h")
xlabel!("x, cm")
ylabel!("T(x), °C")
```
